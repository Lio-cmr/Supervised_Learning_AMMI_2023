```python
def inner(x,y):
  

  """
	r = inner(x,y)
  Compute the inner product of two vectors x and y.

  Parameters
  ----------
  x: vector of size n
  y: vector of size n

  Return
  ------
  r: a float
  
  Example
  -------
  >>> x= np.array([-8, 4, 6, 2, 5])
  >>> y= np.array([2, 2, -1, 4, 3])
  >>> r= inner(x,y)
  >>> r
  9.0
  
  """
##### WRITE YOUR CODE HERE #####

  if (len(x) != len(y)):
    return False
  prod = 0
  for i in range(len(x)):
    prod += x[i]*y[i]
  return prod
  
```

```python
def hadamard(x,y):

  """
	r = hadamard(x,y)
  Compute the hadamard (element-wise) product of two vectors x and y.

  Parameters
  ----------
  x: vector of size n
  y: vector of size n

  Return
  ------
  r: an array of length n
  
  Example
  -------
  >>> x= np.array([-8, 4, 6, 2, 5])
  >>> y= np.array([2, 2, -1, 4, 3])
  >>> r= hadamard(x,y)
  >>> r
  array([-16.,   8.,  -6.,   8.,  15.])
  """
  ##### WRITE YOUR CODE HERE #####
  return [x[i]*y[i] for i in range(len(x)) if len(x) == len(y)]
  pass
```

```python
def outer(x,y):

  """
	r = outer(x,y)
  Compute the inner product of two vectors x and y.

  Parameters
  ----------
  x: vector of size m
  y: vector of size n

  Return
  ------
  M: an mxn matrix 
  
  Example
  -------
  >>> x= np.array([8, 4, 6, 2, 5])
  >>> y= np.array([2, 2, 1, 4, 3, 9, 10])
  >>> M= outer(x,y)
  >>> M
  array([[16. 16.  8. 32. 24. 72. 80.]
  [ 8.  8.  4. 16. 12. 36. 40.]
  [12. 12.  6. 24. 18. 54. 60.]
  [ 4.  4.  2.  8.  6. 18. 20.]
  [10. 10.  5. 20. 15. 45. 50.]])
  
  """
  ##### WRITE YOUR CODE HERE #####
  M = np.zeros((len(x), len(y)))
  for i in range(len(x)):
    for j in range(len(y)):
      M[i, j] = x[i]*y[j]
  return M
  pass
```

```python
def normalize(x):
  
  """
	s = mynormalize(x)
   Return a vector s that is the normalized vector x.
   
  Parameters
  ----------
   x: a vector of size n
   
  Return
  ------
  s: a normalized vector of size n
  
  Example
  -------
  >>> x= np.array([-8,4,6, 2, 5])
  >>> s= normalize(x)
  >>> s
  array([-0.66436384  0.33218192  0.49827288  0.16609096  0.4152274 ])

  """
  ##### WRITE YOUR CODE HERE #####
  x_had = np.zeros((len(x)))
  for i in range(len(x)):
    x_had[i] = x[i] / mynorm(x,2)
  return x_had

  pass
```

```python
def projection(x,y):

  """
	 s = project(x, y)
   Project vector x onto vector y.

   Parameters
   ----------
   x: a vector of size n
   y: a vector of size n

   Return
   ------
   s: an arrray of size n
   
   Example
   -------
   >>> x= np.array([8,4,6, 2, 5])
   >>> y= np.array([2,2,1, 4,3])
   >>> s= projection(x,y)
   >>> s
   array([ 0.52941176  0.52941176 -0.26470588  1.05882353  0.79411765])
   
  """
  ##### WRITE YOUR CODE HERE #####
  projec = np.zeros((len(y)))
  scal = inner(y, x) / (mynorm(y,2) ** 2)
  for j in range(len(y)):
    projec[j] = scal * y[j]
  return projec
    
```

```python
def cosine_similarity(x,y):
  
  """
	 s = cosine_similarity(x, y)
   cosine similarity between vector x and vector y.

   Parameters
   ----------
   x: a list of size n
   y: a list of size n

   Return
   ------
   s: a float
   
   Example
   -------
   >>> x= np.array([8,4,6, 2, 5])
   >>> y= np.array([2,2,1, 4,3])
   >>> r= cosine_similarity(x,y)
   >>> r
   0.12817964067657414
   
  """
  ##### WRITE YOUR CODE HERE #####
  cos_theta = inner(x, y) / (mynorm(x,2) * mynorm(y,2))
  return cos_theta

  pass
```

```python
def matrix_norm(A, p):
    """
      p_norm_A = matrix_norm(A, p)
      p-norm of a matrix A.

      Parameters
      ----------
      A: a nxn matrix.
      p: int or string. This parameter determine the norm type.
      
      Return
      ------
      p_norm_A: a float
      
      Example
      -------
      >>> A= np.array([[5,-4,2], [-1,2,3],[-2,1,0]])
      >>> p_norm_A= matrix_norm(A,p= 1) # p may be: "inf", 1 or F
      >>> p_norm_A
      10.0 
      
    """
    ##### WRITE YOUR CODE HERE #####
    if p  == 1:
      S = np.zeros(A.shape[1])
      for i in range(A.shape[1]):
        S[i] = mynorm(A[:,i], 1)
      return max_of_an_array(S)

    elif p == "inf":
      M = np.zeros(A.shape[0])
      for i in range(A.shape[0]):
        M[i] = mynorm(A[i,:], 1)
      return max_of_an_array(M)

    elif p == "F":
      som = 0
      for i in range(A.shape[0]):
        for j in range(A.shape[1]):
          som +=  A[i,j]*A[i][j]
      return som **(1/2) 
      


A= np.array([[5,-4,2], [-1,2,3],[-2,1,0]])

matrix_norm(A, 1)
```
