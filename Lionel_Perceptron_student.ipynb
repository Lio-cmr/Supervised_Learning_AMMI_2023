## Combines the dataset and Shuffle them using numpy

```python
### Combines the dataset and Shuffle them

dataset = np.hstack((X,y.reshape(-1,1)))
```

```python
dataset[0]
```

```python
n,_=dataset.shape
```

```python
#### Shuffle the dataset and split
perm_index = np.random.permutation(n)

dataset_ = dataset[perm_index]

p = 0.8

X_train, Y_train = dataset_[:round(n*p), :-1], dataset_[:round(n*p), -1]
X_test, Y_test = dataset_[round(n*p):, :-1], dataset_[round(n*p):, -1]
```

```python
print(X_train.shape, X_test.shape)
```

```python
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
class perceptron:
    
    eps = 1e-5
    def __init__(self,theta,iter_= 2):
        self.iter_ = iter_
        self.theta = theta
        
        
    def train(self, X, y):
        n,_= X.shape
        ## your code here
        for j in range(self.iter_):
          for i in range(n):
            ypred = np.sign(self.theta.T @ X[i])
            if (ypred * y[i] < 0):
              self.theta = self.theta + y[i]*X[i]
            else:
              self.theta = self.theta
        return self.theta
        
                    
    def train_(self, X, y):
        n,_= X.shape
        while True:
          ## add your code here
          self.theta1 = self.theta
          if np.linalg.norm(self.theta - self.theta1) < self.__class__.eps:
            break
        return self.theta  
    
    def predict(self,X):
      return np.sign(X@self.theta)
    
    def accuracy(self,X,Y):
      y_pred = self.predict(X)
      result = np.mean(y_pred == Y)
      return result*100
      

            
```

## Instanciate a new perceptron model and train it

```python
theta = np.random.rand(2)*1e-4
perc = perceptron(theta)
```

```python
perc.theta
```

```python
perc.train(X_train, Y_train)
```

```python
perc.theta
```

```python
slope = - perc.theta[0]/perc.theta[1]
min = np.amin(X_train[:, 0])
max = np.amax(X_train[:, 0])

x1 = [min, max]
x2 = [slope*min , slope*max]
plt.scatter(X_train[:,0], X_train[:, 1], marker="o", c=Y_train)
plt.plot(x1, x2)
```
