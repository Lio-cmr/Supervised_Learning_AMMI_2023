## Forward and Backward substitution and Gaussian Elimination

```python
def backsubs(U, b):

  """
  x = backsubs(U, b)
  Apply back substitution for the square upper triangular system Ux=b. 

  Parameters
  -------
    U: nxn square upper triangular array
    b: n array
    

  Returns
  -------
    x: nx1 array
  """
  ##### WRITE YOUR CODE HERE #####
  n = len(U)
  x = np.zeros_like(b, dtype=float)
  for i in range(n - 1, -1, -1):
    x[i] += b[i] / U[i,i]
    for j in range(n-1, i, -1):
      x[i] += (- (U[i, j] * x[j])) / U[i,i]
  return x
  


def forwardsubs(L, b):
  


  """
  	x = forsubs(L, b)
    Apply forward substitution for the square lower triangular system Lx=b. 

    Parameters
    -------
      L: nxn square lower triangular array
      b: n array
      
    Returns
    -------
      x: n array

  """
  ##### WRITE YOUR CODE HERE #####
  n = len(b)
  x = []
  for i in range(n):
    x.append(b[i])
    for j in range(i):
      x[i]=x[i]-(L[i, j]*x[j])
      x[i] = x[i]/L[i, i]
  return x

  



  
def fowardElimination(A,b):
  # Naive Gaussian Elimination (with no pivoting) also called Forward Elimination
  """
  	U,b_new = fowardElimination(A, b)
    Apply fowardElimination for the square system Ax=b. 

    Parameters
    -------
      A: nxn square matrix 
      b: n array

    Returns
    -------
      U: nxn upper triangular matrix
      b_new: n array
  """
  ##### WRITE YOUR CODE HERE #####
  n = len(A)
  for i in range(0, n-1):
    for j in range(i+1, n):
      A[j] = A[j] - A[i]/A[i,i]*A[j,i]
      b[j] = b[j] - b[i]/A[i,i]*A[j,i]
  U = A
  b_new = b    
    
  return U, b_new

  

def ge(A,b):
  
  """
  	x = ge(A, b)

    Parameters
    -------
      A: nxn square matrix 
      b: n array

    Returns
    -------
      x: n-array

  """
  ##### WRITE YOUR CODE HERE #####
  # Call fowardElimination
  U,b_new = fowardElimination(A, b)
  # Call backsubs
  return backsubs(U, b_new)
  
```

## LU Decomposition/Factorization

```python
def ludecom(A):
    
  """
    L,U = ludecom(A)
    Apply LU decomposition to mxm square matrix. 

    Parameters
    -------
    A: mxm square matrix 

    Returns
    -------
    L: mxm square matrix 
    U: mxm square matrix 
  """

  ##### WRITE YOUR CODE HERE #####
  n = A.shape[0]
  U = A.copy()
  L = np.eye(n, dtype=np.double)
  for j in range(n):
    factor = U[j+1:, j] / U[j, j]
    L[j+1:, j] = factor
    U[j+1:] = U[j+1:] - factor[:, np.newaxis] * U[j]

  return L, U


def solve(A, b):
    
    """
     x = solve(A, b)
    Apply LU decmposition, forward and backward substitution for the square system Ax=b. 

    Parameters
    -------
    A: mxm square matrix 
    b: m vector
        
    Returns
    -------
    x: n vector
    
    """
    ##### WRITE YOUR CODE HERE #####
    L,U = ludecom(A)
    y = forwardsubs(L, b)
    return  backsubs(U, y)
    

A= np.array([[1,1,1, 4],[2,3,5, 2],[4,6,8, 3], [3,9,2, 1]],dtype= np.float64)
b= np.array([2,3,1, 4],dtype= np.float64)

x= solve(A, b)

print(A@x-b)
```
