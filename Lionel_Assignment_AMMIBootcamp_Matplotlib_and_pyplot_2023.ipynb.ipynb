```python
# data for plottig the figure
x_axis = ["State SAC", "CURL", "Dreamer", "SLACv1", "SAC+AE", "PlaNet", "Pixel SAC"]
env_steps_100k = [779, 560, 296, 437, 351, 137, 133] # Values for the blue bars
env_steps_500k = [923, 914, 795, 767, 745, 406, 188] # Values for the red bars
y_label = "Evaluation Score" 
title = "Median Scores on DMControl100k and DMControl500k"
```

```python
############## Add your code here ##############
'''
General Steps:
1. Make the multibar plot with the required labels, colors, and legend.
2. Make sure the font size is 20.
3. Add the title and the other stylistic modifications.
'''
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
      # we find the height of the bar
        height = rect.get_height()
        # we start the annotation
        plt.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize = 20)


# we create an np.array with the len of x_axis
X_axis = np.arange(len(x_axis))
# we define the bar width
bar_width = 0.38
rects1 = plt.bar(X_axis - 0.22 ,env_steps_100k, width = bar_width, color = 'b', label = 'env_steps_100k')
rects2 = plt.bar(X_axis + 0.22,env_steps_500k,width = bar_width, color = "r", label = 'env_steps_500k')
# we call our auto label function on each container
autolabel(rects1)
autolabel(rects2)
# Add the title
plt.title(title, fontsize = 20)
# Modify the X_axis by x_axis
plt.xticks(X_axis, x_axis, fontsize = 20)
plt.yticks(fontsize = 20)
# Define a limit in the y axis
plt.ylim(0, 1000)
# Add the legend
plt.legend(fontsize = 20)
# Add the label on the y axis
plt.ylabel(y_label, fontsize = 20)


################################################
plt.show()
```

#Figure Two, Training curve
Plot the figure below

- The x range is from 0, to 100, with a step size of 1
- The function for the blue curve is exp(-0.01 * x)
- The function for the red curve is 0.1 + exp(-0.1*x)
- The black dashed line is just a constant with a value of 0.4


Set the font size to 20, color values are "tab:blue", "tab:red", and "black", if there are other values that you need to set, choose any appropriate values.



training.png

```python
############## Add your code here ##############
'''
General Steps:
1. Define the domain x.
2. Define the function values for each curve.
3. Make any changes you think are needed. 
'''
# we define x
x = np.arange(0, 100, 1)
# we define the 3 models
y_1 = np.exp(-0.01*x)
y_2 = 0.1 + np.exp(-0.1*x)
baseline = np.array([0.4]*len(x))

plt.plot(x, y_1, label = 'Model1', color = 'b')
plt.plot(x, y_2, label = 'Model2', color = 'r')
plt.plot(x, baseline, linestyle = 'dotted', color = 'black', label = 'Baseline')
# Add the title
plt.title('Training Curve',  fontsize = 20)
# Add the label on the x axis
plt.xlabel('Training Iteration',  fontsize = 20)
# Add the label on the y axis
plt.ylabel('Training error',  fontsize = 20)
# Add the legend
plt.legend( fontsize = 20)
# Set the fontsize on the x axis
plt.xticks(fontsize = 20)
# Set the fontsize on the y axis
plt.yticks(fontsize = 20)


################################################
plt.show()
```

```python
############## Helper function to generate some radnom data from a gaussiain ##############
def generte_data(mean_x1, mean_x2, std, size):
  x1 = mean_x1 + std * np.random.randn(size)
  x2 = mean_x2 + std * np.random.randn(size)
  return x1, x2
###########################################################################################

# generate data for a scatter plot
x1, y1 = generte_data(0.2, 0.5, 0.1, 1000)
x2, y2 = generte_data(0.5, -0.5, 0.1, 1000)
x3, y3 = generte_data(0, 0, 0.1, 1000)

############## Add your code here ##############
'''
General Steps:
1. Generate the scatter plot for the data provided above.
2. Add the title, x and y labels.
3. Modify the legends, x and y ticks to match the plot above.
'''
plt.scatter(x1, y1, label = 'Class1', color = 'b')
plt.scatter(x2, y2, label = 'Class2', color = 'g', marker = '+')
plt.scatter(x3, y3, color = 'r', label = 'Class3', marker ="^")
# Add the title
plt.title('Dummy Classification Problem',  fontsize = 20)
# Add the label on the x axis
plt.xlabel('X',  fontsize = 20)
# Add the label on the x axis
plt.ylabel('Y',  fontsize = 20)
# Add the legend
plt.legend( fontsize = 20)
# Set the fontsize on the x axis
plt.xticks(fontsize = 20)
# Set the fontsize on the y axis
plt.yticks(fontsize = 20)


################################################
plt.show()
```
